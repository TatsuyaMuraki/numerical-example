# 深さを変化させると

```python
#ライブラリのインポート
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt


#トレーニングデータとテストデータの作成
x_train = np.linspace(-3,3,10).reshape(-1,1)
y_train = x_train**2

x_real = np.linspace(-3,3,100).reshape(-1,1)
y_real = x_real**2


#グラフの描画
for i in range(1,10):    
    tree = DecisionTreeRegressor(max_depth=i)
    tree.fit(x_train, y_train)
    y_pred = tree.predict(x_real)
    plt.subplot(3,3,i)
    plt.plot(x_real, y_real)
    plt.plot(x_real,y_pred)
    plt.scatter(x_train, y_train)
```

```python
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file = 'reg.dot')
```

# 決定木の可視化

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

#データの取得
iris = datasets.load_iris()
X = iris.data[:,[2,3]]
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

#モデルの構築
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)
tree.fit(X_train, y_train)

#可視化のdotファイルを作成
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file = 'tree.dot',feature_names=['petal length','petal width'])

#dotファイルをpngファイルに変換（コマンドライン）
'''
dot -Tpng tree.dot tree.png
'''

#変数重要度を出力
importances = tree.feature_importances_
indices = np.argsort(importances)
plt.barh(range(len(importances)), importances, color='b', align='center')
plt.yticks(range(2) ,['petal length','petal width'])
```

# 決定木の数値例を作ってみる

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

#変数の数
p = 5

#関数の決定
def func(x):
        y= np.sin(x[1]) + x[3]
        return y

#データの取得
#トレーニングデータの作成　-1<x<1になっている
N = 30
X = []
Y = []
for i in range(N):    
    x = 2*np.random.rand(p) -1
    x = x.tolist()
    X.append(x)
    y = func(X[i]) + 0.05*np.random.rand(1)
    y = y.tolist()
    Y.append(y)
X = np.array(X)   
Y = np.array(Y)     

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

#モデルの構築
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(max_depth=3, random_state=0)
tree.fit(X_train, y_train)



#可視化のdotファイルを作成
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file = 'tree.dot',feature_names=['1','2','3','4','5'])



#dotファイルをpngファイルに変換（コマンドライン）
'''
dot -Tpng tree.dot tree.png
'''

#変数重要度を出力
importances = tree.feature_importances_
indices = np.argsort(importances)
plt.barh(range(len(importances)), importances, color='b', align='center')
plt.yticks(range(5),['1','2','3','4','5'] )
```

# 決定木をランダムフォレストに変えてみる

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

#変数の数
p = 5

#関数の決定
def func(x):
        y= np.sin(x[1]) + x[3]
        return y

#データの取得
#トレーニングデータの作成　-1<x<1になっている
N = 30
X = []
Y = []
for i in range(N):    
    x = 2*np.random.rand(p) -1
    x = x.tolist()
    X.append(x)
    y = func(X[i]) + 0.05*np.random.rand(1)
    y = y.tolist()
    Y.append(y)
X = np.array(X)   
Y = np.array(Y) .reshape(-1,)    

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

#モデルの構築
from sklearn.ensemble import RandomForestRegressor
tree = RandomForestRegressor(n_estimators=1000, max_depth=3, random_state=0)
tree.fit(X_train, y_train)


'''
#可視化のdotファイルを作成
from sklearn.tree import export_graphviz
export_graphviz(tree, out_file = 'tree.dot',feature_names=['1','2','3','4','5'])
'''


#dotファイルをpngファイルに変換（コマンドライン）
'''
dot -Tpng tree.dot tree.png
'''

#変数重要度を出力
importances = tree.feature_importances_
indices = np.argsort(importances)
plt.barh(range(len(importances)), importances, color='b', align='center')
plt.yticks(range(5),['0','1','2','3','4'] )
```
