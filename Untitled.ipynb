# sklernでやってみる

```python
from sklearn.gaussian_process import kernels as sk_kern
from sklearn.gaussian_process import GaussianProcessRegressor

kernel = sk_kern.RBF(1.0, (1e-3, 1e3))
clf = GaussianProcessRegressor(
    kernel=kernel,
    alpha=1e-10,
    normalize_y=True)


np.random.seed(123)
x = np.linspace(-5,5,200)
y = np.sin(x)
plt.plot(x, y)


N = 2


x_train = 10*np.random.rand(N) -5
y_train = np.sin(x_train)
plt.plot(x, y)
plt.scatter(x_train, y_train)

x_test = np.linspace(-5,5., 200).reshape(-1, 1)
    


for i in range(1,1+len(x_train)):
    clf.fit(x_train[:i].reshape(-1, 1), y_train[:i])
    clf.kernel_
    mean, std= clf.predict(x_test, return_std=True)

    plt.plot(x_test[:, 0], mean, color="C0", label="predict mean")
    plt.fill_between(x_test[:, 0], mean + std, mean - std, color="C0", alpha=.3,label= "3 sigma confidence")
    plt.plot(x_train, y_train, "o",label= "training data")
    plt.title("Scikit-learn")
    plt.legend()
```
