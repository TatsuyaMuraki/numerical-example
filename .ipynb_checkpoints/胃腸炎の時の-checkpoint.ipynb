```python
x = np.linspace(-5,5,100)
y = np.sin(x)
```

```python
kernel = GPy.kern.RBF(input_dim=1, variance=1, lengthscale=0.2)
```

```python
model = GPy.models.GPRegression(x.reshape(-1,1), y.reshape(-1,1), kernel = kernel)
```

```python
model.plot()
```

```python
import GPy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


kernels = []
kernels.append(GPy.kern.RBF(1))
kernels.append(GPy.kern.Linear(1))
kernels.append(GPy.kern.PeriodicExponential(1))

np.random.seed(123)
```

```python
kernel1 = GPy.kern.RBF(1)
kernel2 = GPy.kern.RBF(1) + GPy.kern.Bias(1)
kernel3 = GPy.kern.Bias(1)
```

```python
kernel3.plot()
```

```python
np.random.seed(12)
x_real = np.linspace(-5,5,100)
y_real = 3 * x_real
plt.plot(x_real, y_real)

x_train = 10 * np.random.rand(10) - 5 
y_train = 3 * x_train + np.random.randn(10)
plt.scatter(x_train, y_train, color='red')

x_test = 10 * np.random.rand(10) - 5 
y_test
plt.scatter(x_test, y_test, color='green')

model1 = GPy.models.GPRegression(x_train.reshape(-1,1), y_train.reshape(-1,1), kernel=kernel1)
model2 = GPy.models.GPRegression(x_train.reshape(-1,1), y_train.reshape(-1,1), kernel=kernel2)
model1.optimize()
model2.optimize()
model1.plot()
model2.plot()

y_pred = model1.predict(x_test.reshape(-1,1))
plt.scatter(x_test.reshape(-1,1), y_pred[0], color='blue')





#for i in range(len(x_test)):
    

```
